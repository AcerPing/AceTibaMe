```
import pandas as pd
train_df = pd.read_csv("train.csv",encoding="utf-8")
test_df = pd.read_csv("test.csv",encoding="utf-8")
test_df
```

```
#SalePrice → y_train(答案)
train_df_drop = train_df.drop(["SalePrice"], axis=1)
datas = pd.concat([train_df_drop, test_df], axis=0)
datas = datas.drop(["Id"], axis=1)
datas
```

```
na = datas.isna().sum()
na[na != 0].sort_values(ascending=False)
```

```
datas = datas.drop(na[na > 1000].index, axis=1)
datas
```

```
#1.填缺失值:有大小關係的填中位數
med = train_df.median().drop(["MSSubClass"])
train_df = train_df.fillna(med)
train_df
```

```
# One-Hot Encoding
datas = pd.get_dummies(datas)
datas = pd.get_dummies(datas, columns=["MSSubClass"])
datas
```

```

# 取列: .iloc -> [第一筆資料, 第二筆資料...]
x_train = datas.iloc[:train_df.shape[0]]
y_train = train_df["SalePrice"]
x_predict = datas.iloc[train_df.shape[0]:]
predict_id = test_df["Id"]
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
clf = RandomForestRegressor(random_state=0)
params = {"n_estimators":range(20,100),
      "max_depth":range(5,10)}
search = GridSearchCV(clf,params,cv=10,n_jobs=-1,scoring="r2")
search.fit(x_train,y_train)
print(search.best_params_)
print(search.best_score_)
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import numpy as np
clf = RandomForestRegressor(n_estimators=77,max_depth=7,random_state=0)
scores = cross_val_score(clf,x_train,y_train,cv=10,n_jobs=-1,scoring=="r2")
print(scores)
print(np.average(scores))
```
