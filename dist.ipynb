```
import numpy as np
import seaborn as sns
%matplotlib inline
np.random.normal()
samples = np.random.normal(loc=10,scale=2,size=10000)
sns.distplot(samples)
```

```
s = samples[samples > 0]+1
sns.distplot(s)
```

```
from scipy.stats import skew
new = s ** 5
sns.distplot(new)
print("skew:",skew(new))
```

```
from scipy.stats import boxcox_normmax,boxcox
lamda = boxcox_normmax(new)
print(lamda)
new2 = boxcox(new,lamda)
sns.distplot(new2)
print(skew(new2))
```

```
new = s ** (1/5)
sns.distplot(new)
print("skew:",skew(new))
```

```
lamda = boxcox_normmax(new)
print(lamda)
new2 = boxcox(new,lamda)
sns.distplot(new2)
print(skew(new2))
```
